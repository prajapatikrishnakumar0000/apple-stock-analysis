```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"..\dataset\raw_data_1\sales_data.csv")
```

```python
df
```

```python
df.columns
```

```python
##rename Price to date
df = df.rename(columns={"Price": "Date"})
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
##delete the satrting 2 row , coz its useless
df = df.drop(index=[0, 1]).reset_index(drop=True)
```

```python
df
```

```python
df.dtypes
```

```python
##as datatype are not suitable for ploting so convert into int or float
df["Date"] = pd.to_datetime(df["Date"])

numeric_column = ["Close", "High", "Low", "Open", "Volume"]
df[numeric_column] = df[numeric_column].astype(float)
```

```python
df.dtypes
```

```python
## sort date sequencely may be in between it might be different

df = df.sort_values("Date").reset_index(drop=True)
```

```python
df
```

```python
df.isnull().sum()  ##removed earlyier in row 0 & 1
```

```python
## think of that price < 0 , high < low...etc for this we do sanity check . it should come 0

(df[["Close", "High", "Low", "Open", "Volume"]] <= 0).sum()
```

```python
df = df.set_index("Date")
```
